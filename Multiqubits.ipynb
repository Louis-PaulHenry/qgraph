It seems like the multi-qubit encoding holds pretty well even in systems of multiple qubits. We've made a test with a system composed of a chain of three qubits with nn blockade under a random pulse. On one hand we simulated the system where each logical qubit is encoded by three physical qubits, and on the other hand we simulated the system with three logical and physical qubits under the same pulse but with Rabi frequency scaled by a factor of $\sqrt 3$. 

The picture occupations.png shows four expectation values (We will add the suffix OQE and TQE to refer to one-qubit and three-qubit encoding respectively): $<n_0>_{OQE}$ corresponds pretty well to $<n_0+n_1+n_2>_{TQE}$, and $<n_1>_{OQE}$ corresponds pretty well to $<n_3+n_4+n_5>_{TQE}$.
Tests with different random pulses have a similar behaviour.
